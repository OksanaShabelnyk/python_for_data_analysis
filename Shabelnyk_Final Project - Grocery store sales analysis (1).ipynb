Прослідкувати динаміку зміни Gross Bill Amount з залежності від категорії товарів та років, коли здійснювалися трансакції.

```python
pivot_sub_gros = pd.pivot_table(data = df, index = ['Year'], columns=['Category'], values=['Gross Bill Amount'], 
                                    aggfunc='sum')
pivot_sub_gros
```

```python
pivot_sub_gros.T.plot();
plt.xticks(rotation=90);
plt.grid(axis='both', alpha=.3);
plt.show();
```
